```python
import tensorflow as tf
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
(train_images, train_lables),(test_images,test_labels) = tf.keras.datasets.fashion_mnist.load_data()
```

```python
train_images,test_images = train_images/255.0,test_images/255.0
```

```python
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(test_images[i], cmap=plt.cm.binary)
    plt.xlabel(class_names[train_lables[i]])
```

```python
model = tf.keras.Sequential()
```

```python
model.add(tf.keras.layers.Flatten(input_shape=(28, 28)))
```

```python
model.add(tf.keras.layers.Dense(128, activation=tf.nn.relu))
```

```python
model.add(tf.keras.layers.Dense(16, activation=tf.nn.relu))
```

```python
model.add(tf.keras.layers.Dense(10,activation=tf.nn.softmax))
```

```python
model.compile(optimizer=tf.train.AdamOptimizer(),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(train_images,train_lables,epochs=5)
```

```python
prediction = model.predict(test_images)
```

```python
result = np.equal([np.argmax(x) for x in prediction],test_labels)
```

```python
sum(result)*1.0/len(result)
```
